```python
import pyfiglet

fonts = pyfiglet.FigletFont.getFonts()

print("Available Fonts (showing few):")
print(fonts[:10])   # show only first 10 to avoid long list

text = input("\nEnter text: ")
font = input("Enter font name (example: slant, block, banner): ")

if font not in fonts:
    print("Font not found! Using default.")
    font = "standard"

ascii_art = pyfiglet.figlet_format(text, font=font)
print("\nASCII Art Output:\n")
print(ascii_art)
```
